```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as tck
import numpy as np
import pandas as pd
import os

sns.set_theme()
```

```python
file = "gpu20000.csv"
nIter = 5

data = pd.read_csv(file,header=0,sep=";")
init = data["init"]
initComm = data["initComm"]
gather = data["gather"]
resAlloc = data["resAlloc"]
dGemm = data["dGemm"]
place = data["place"]
mult = data["mult"]
total = data["total"]
initMeans = [init[i:i+nIter].mean() for i in range(0, len(init), nIter)]
initCommMeans = [initComm[i:i+nIter].mean() for i in range(0, len(initComm), nIter)]
gatherMeans = [gather[i:i+nIter].mean() for i in range(0, len(gather), nIter)]
resAllocMeans = [resAlloc[i:i+nIter].mean() for i in range(0, len(resAlloc), nIter)]
dGemmMeans = [dGemm[i:i+nIter].mean() for i in range(0, len(dGemm), nIter)]
placeMeans = [place[i:i+nIter].mean() for i in range(0, len(place), nIter)]
multMeans = [mult[i:i+nIter].mean() for i in range(0, len(mult), nIter)]
totalMeans = [total[i:i+nIter].mean() for i in range(0, len(total), nIter)]
dataLists = [initMeans, initCommMeans, gatherMeans, resAllocMeans, dGemmMeans, placeMeans, multMeans, totalMeans]
xLabels = [2**i for i in range(5)]
for i in range(len(dataLists)):
    plt.plot(xLabels, dataLists[i], label=data.columns[i])
plt.xlabel("Number of nodes")
plt.ylabel("Time (s)")
plt.legend()
plt.plot()
```
